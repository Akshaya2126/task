```python
def calculator(a,b):
    x=int(input("Enter your Choice"))
    if x==1:
        add=a+b
        print(add)
    elif x==2:
        sub=a-b
        print(sub)
    elif x==3:
        mul=a*b
        print(mul)
    elif x==4:
        div=a/b
        print(div)
    elif x==5:
        mod=a//b
        print(mod)
    elif x==6:
        exp=a**2
        print(exp)
    elif x==7:
        exit

a=int(input("Enter the number 1:"))
b=int(input("Enter the number 2:"))
c=calculator(a,b)
```
